# Assignment: Segmenting and Clustering Neighborhoods in Toronto

## 1- Import Postal Codes from URL in Table1

```python
import pandas
Data = pandas.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
PostalCodesData = Data[0]
PostalCodesData
```

## 2- Remove Boroughs are not assigned

```python
PostalCodesData = PostalCodesData.loc[lambda x: x.Borough != 'Not assigned']
PostalCodesData.reset_index(drop=True,inplace=True)
PostalCodesData.head(10)
```

## 3- Replace Neighbourhoods are not assigned with their Borough

```python
PostalCodesData.loc[lambda x: x.Neighbourhood == 'Not assigned']
```

#### There are not any Neighbourhood with Not assigned value to replace.

## 4- Group neighbourhoods by Postal Codes

```python
PostalCodesData = PostalCodesData.groupby('Postal Code',as_index=False).agg(lambda x: ','.join(set(x.dropna())))
PostalCodesData.head(20)
```
