# 1- Data Gathering

### Import data of top Nineteen tourists attractions in Paris is ranked by tripadvisor.com

```python
import pandas
Attractions = pandas.read_excel("https://github.com/Poria-GitHub/IBM-Data-Science-Professional-Certificate/raw/main/Part%204%20-%20Data%20-%20Twenty%20Top%20Things%20To%20Do%20in%20Paris.xlsx")
Attractions.head()
```

### Extract geographical coordinates for each tourist attractions 

```python
from geopy.geocoders import Nominatim
Geo = Nominatim(user_agent= 'ny_explorer')
Temp= {'Name':[],'Latitude':[],'Longitude':[]}
for Adress in Attractions['Name']:
    Location = Geo.geocode(Adress)
    Temp['Name'].append(Adress)
    Temp['Latitude'].append(Location.latitude)
    Temp['Longitude'].append(Location.longitude)
AttractionsData= pandas.merge(left=Attractions,right=pandas.DataFrame(Temp),on='Name')
AttractionsData = round(AttractionsData,4)
AttractionsData
```

### Import data of top Twenty top restaurants in Paris is ranked by tripadvisor.com and extract geographical coordinates

```python
Restaurants = pandas.read_excel("https://github.com/Poria-GitHub/IBM-Data-Science-Professional-Certificate/raw/main/Part%204%20-%20Data%20-%20Restaurants%20in%20Paris.xlsx")
Temp= {'Name':[],'Latitude':[],'Longitude':[]}
for Adress in Restaurants['Name']:
    Location = Geo.geocode(Adress)
    Temp['Name'].append(Adress)
    Temp['Latitude'].append(Location.latitude)
    Temp['Longitude'].append(Location.longitude)
RestaurantsData= pandas.merge(left=Restaurants,right=pandas.DataFrame(Temp),on='Name')
RestaurantsData = round(RestaurantsData,4)
RestaurantsData
```

### Import data of top Five shopping Malls in Paris is ranked by france-hotel-guide.com and extract geographical coordinates

```python
ShoppingMalls = pandas.read_excel("https://github.com/Poria-GitHub/IBM-Data-Science-Professional-Certificate/raw/main/Part%204%20-%20Data%20-%20Shopping%20Malls%20in%20Paris.xlsx")
Temp= {'Name':[],'Latitude':[],'Longitude':[]}
for Adress in ShoppingMalls['Name']:
    Location = Geo.geocode(Adress)
    Temp['Name'].append(Adress)
    Temp['Latitude'].append(Location.latitude)
    Temp['Longitude'].append(Location.longitude)
ShoppingMallsData= pandas.merge(left=ShoppingMalls,right=pandas.DataFrame(Temp),on='Name')
ShoppingMallsData = round(ShoppingMallsData,4)
ShoppingMallsData
```

# 2- Data Preparing

### Scale data by Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
Data = HotelsData.drop(['Name','Id','Latitude','Longitude','Rate'],1)
scaler = StandardScaler()
scaler.fit(Data)
DataScale = pandas.DataFrame(scaler.transform(Data))
DataScale.columns=Data.columns
DataScale
```

# 3- Clustering

### First, find the best number of clusters

```python
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
import numpy
distortions=[]
Clusters=range(1,10)
for k in Clusters: 
    #Building and fitting the model 
    Model = KMeans(n_clusters=k).fit(DataScale)
    distortions.append(sum(numpy.min(cdist(DataScale, Model.cluster_centers_,'euclidean'),axis=1)) / DataScale.shape[0])

import matplotlib.pyplot  as plt
plt.plot(Clusters, distortions, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Distortion') 
plt.title('The Elbow Method using Distortion') 
plt.show()
```

### Then, examine cluster of each hotels

```python
BestNumber=4 #Best number of clusters
KMeansResult = KMeans(n_clusters=BestNumber, random_state=0).fit(DataScale)
print("Hotels' Cluster = \n" ,KMeansResult.labels_)
HotelCluster = pandas.DataFrame(KMeansResult.labels_)
HotelsData = HotelCluster.join(HotelsData).rename(columns={0:'Cluster Number'})
HotelsData
```
